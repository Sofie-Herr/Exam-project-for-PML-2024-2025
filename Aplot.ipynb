```python
import json


with open('fid_data.json', 'r') as f:
    fid_data = json.load(f)

with open('fid_data_SDE.json', 'r') as f:
    fid_data_SDE = json.load(f)

print("FID Data:", fid_data)
print("FID Data SDE:", fid_data_SDE)
```

```python
import matplotlib.pyplot as plt
```

```python
epochs_fid = fid_data['epochs']
scores_fid = fid_data['fid_scores']

epochs_fid_SDE = fid_data_SDE['epochs']
scores_fid_SDE = fid_data_SDE['fid_scores']

# Create the plot
plt.figure(figsize=(12, 6))

plt.plot(epochs_fid, scores_fid, label='FID Data', marker='o')
plt.plot(epochs_fid_SDE, scores_fid_SDE, label='FID Data SDE', marker='x')

# Customize the plot
plt.title('FID Scores Comparison Over Epochs', fontsize=16)
plt.xlabel('Epochs', fontsize=14)
plt.ylabel('FID Scores', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.ylim(0,2000)

# Show the plot
plt.tight_layout()
plt.show()
```
